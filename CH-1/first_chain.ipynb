# **first chain**

```python
import os
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
from typing import TypedDict
from langchain_core.messages import HumanMessage,SystemMessage,AIMessage
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

load_dotenv()

llm = ChatOpenAI(model="gpt-5-mini", temperature=0)

prompt_template = ChatPromptTemplate.from_messages([
    ("system","You are a helpful assistant"),
    ("human","{input}")
])

str_parser = StrOutputParser()

#template = prompt_template.invoke({"What is capital of france?"})

#es = llm.invoke(template)

##final_result = res.content

#print(final_result)

# using chain we can invoke them all one after another

chain = prompt_template | llm | str_parser

chain.invoke({"input":"What is the capital of India?"})




```
